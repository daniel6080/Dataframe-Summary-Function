```python
import numpy as np
import pandas as pd
```

### Prompt

```python
test=input('Please enter the name of the test data file: (e.g. boston_tst.csv): ')
```

```python
df_train=pd.read_csv(train)
df_train
```

```python
df_test=pd.read_csv(test)
df_test
```

### Multiple Linear Regression

#### Train data

```python
Y_train = df_train['medv'] #response variable from train data
Y_train
```

```python
Z_train_col = df_train.columns.delete(12) #columns of predictor variables
Z_train=df_train[Z_train_col] #predictor variables
ones=pd.DataFrame(np.ones(len(df_train),dtype=np.int8))
Z_train=pd.concat([ones,Z_train],axis=1)
Z_train=Z_train.T.reset_index(drop=True).T #reset columns
Z_train #predictor variables with constant term
```

```python
B_train_hat=pd.DataFrame((np.linalg.inv(Z_train.T.dot(Z_train))).dot(Z_train.T).dot(Y_train))
B_train_hat=round(B_train_hat,3)
B_train_hat
```

```python
#Fitted values for train data
Y_train_hat = Z_train.dot(B_train_hat)
Y_train_hat
```

#### Test data

```python
Y_test = df_test['medv'] #response variable from test data
Y_test
```

```python
Z_test_col = df_test.columns.delete(12) #columns of predictor variables
Z_test=df_test[Z_test_col] #predictor variables
ones=pd.DataFrame(np.ones(len(df_test),dtype=np.int8))
Z_test=pd.concat([ones,Z_test],axis=1)
Z_test=Z_test.T.reset_index(drop=True).T #reset columns
Z_test #predictor variables with constant term
```

```python
#Fitted values for test data
Y_test_hat = Z_test.dot(B_train_hat)
Y_test_hat
```

### Model Summary

```python
#Calculate Model R^2
SSE_train = sum((Y_train-Y_train_hat[0])**2)
SST_train = sum((Y_train-Y_train.mean())**2)
R_square_model = round(1-SSE_train/SST_train,4)
R_square_model
```

```python
#Calculate Model MSE
n_train = len(df_train)
p_train = len(Z_train.columns)
MSE_model = round(SSE_train/(n_train-p_train),3)
MSE_model
```

### Prediction Performance

```python
#Calculate Predictive R-square
SSE_test = sum((Y_test-Y_test_hat[0])**2)
SST_test = sum((Y_test-Y_test.mean())**2)
R_square_pred = round(1-SSE_test/SST_test,4)
R_square_pred
```

```python
#Calculate MAE
n_test = len(df_test)
MAE = round(sum(np.abs(Y_test-Y_test_hat[0]))/n_test,3)
MAE
```

```python
#Calculate MAPE
MAPE = round(1/n_test*sum(abs(Y_test-Y_test_hat[0])/abs(Y_test)),3)
MAPE
```

```python
#Calculate RMSE
RMSE = round((SSE_test/n_test)**0.5,3)
RMSE
```

### Write File

```python
f = open("HW2_output.txt",'w')
text = "Coefficients\n-------------\nConstant: "+str(B_train_hat.iloc[0,0])+"\n"
for i in range(1,len(Z_train.columns)):
    text += "Beta"+str(i)+": "+str(B_train_hat.iloc[i,0])+"\n"
text += "\nModel Summary\n-------------\nR-square = " + str(R_square_model) + "\n"
text += "MSE = " + str(MSE_model)
text += "\n\nPrediction Performance\n----------------------\nPredictive R-square = " + str(R_square_pred) + "\n"
text += "MAE = " + str(MAE) + "\n"
text += "MAPE = " + str(MAPE) + "\n"
text += "RMSE = " + str(RMSE)
f.write(text)
f.close()
```
