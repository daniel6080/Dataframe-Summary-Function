```python
import pandas as pd
import numpy as np
```

```python
#1
auto = pd.read_fwf("auto.dat", header = None, widths = [19,4,6,3,2,2,4,5,3,5,4,3,4,4], 
                   na_values='.')
auto.columns = ['model','origin', 'price', 'mpg', 'rep78', 'rep77', 'hroom', 'rseat', 
                'trunk', 'weight', 'length', 'turn', 'displa', 'gratio']
auto
```

```python
auto1 = auto[['price','mpg','hroom','rseat','weight','length','gratio']]
auto1
```

```python
ones=pd.DataFrame({'ones': np.ones(74)})
ones
```

```python
#sample mean (x_bar)
x_bar=ones.transpose().dot(auto1)/len(auto1)
x_bar
```

```python
#compare with built-in function
auto1.mean()
```

```python
ones_matrix=pd.DataFrame(np.ones((74,7)), columns=['price','mpg','hroom','rseat','weight','length','gratio'])
ones_matrix
```

```python
auto2=auto1-x_bar_rep
auto2
```

```python
# sample covariance matrix (S)
S=auto2.transpose().dot(auto2)/(len(auto1)-1)
S
```

```python
#compare with built-in function
auto1.cov()
```

```python
#create matrix with diagonal components repeated for each row
Si=S.copy()
for i in range(7):
    Si.iloc[i,:]=S.iloc[i,i]
Si
```

```python
#create matrix with diagonal components repeated for each column
Sj=S.copy()
for j in range(7):
    Sj.iloc[:,j]=S.iloc[j,j]
Sj
```

```python
#sample correlation matrix (R)
R=S/np.sqrt(Si)/np.sqrt(Sj)
R
```

```python
#2
applicant = pd.read_csv('applican.dat', header = None, delim_whitespace=True)
applicant.columns = ['ID','FL','APP','AA','LA','SC','LC','HON','SMS','EXP',
                     'DRV','AMB','GSP','POT','KJ','SUIT']
applicant
```

```python
#repeat steps in problem 1 to obtain sample correlation matrix of 'applicant' data
#get mean
applicant1 = applicant[['LC','SC','SMS']]
ones=pd.DataFrame({'ones': np.ones(48)})
applicant1_mean=ones.transpose().dot(applicant1)/len(applicant1)
applicant1_mean
```

```python
#get sample covariance matrix
ones_matrix=pd.DataFrame(np.ones((48,3)), columns=['LC','SC','SMS'])
applicant1_mean_rep = pd.concat([applicant1_mean]*48)
applicant1_mean_rep.columns=['LC','SC','SMS']
applicant1_mean_rep.reset_index(inplace = True, drop = True)
applicant2=applicant1-applicant1_mean_rep
applicant1_S=applicant2.transpose().dot(applicant2)/(len(applicant1)-1)
applicant1_S
```

```python
#get sample correlation matrix
Si=applicant1_S.copy()
for i in range(3):
    Si.iloc[i,:]=applicant1_S.iloc[i,i]

Sj=applicant1_S.copy()
for j in range(3):
    Sj.iloc[:,j]=applicant1_S.iloc[j,j]

applicant1_R=applicant1_S/np.sqrt(Si)/np.sqrt(Sj)
applicant1_R
```

```python
from scipy import stats
```

```python
#hypothesis test (LC,SC)
r1=applicant1_R.loc['LC','SC']
test_stat1=r1*np.sqrt(len(applicant1)-2)/np.sqrt(1-r1**2)
test_stat1
```

```python
pval1=stats.t.sf(abs(test_stat1), df=len(applicant1)-2)*2
pval1 #reject H0
```

```python
#hypothesis test (SC,SMS)
r2=applicant1_R.loc['SC','SMS']
test_stat2=r2*np.sqrt(len(applicant1)-2)/np.sqrt(1-r2**2)
test_stat2
```

```python
pval2=stats.t.sf(abs(test_stat2), df=len(applicant1)-2)*2
pval2 #reject H0
```

```python
#hypothesis test (LC,SMS)
r3=applicant1_R.loc['LC','SMS']
test_stat3=r3*np.sqrt(len(applicant1)-2)/np.sqrt(1-r3**2)
test_stat3
```

```python
pval3=stats.t.sf(abs(test_stat3), df=len(applicant1)-2)*2
pval3 #reject H0
```

```python
#3
#Let us find confidence intervals for population correlation at significance level Î±=0.05
#Thus we use the critical value z=1.96 for our calculations
#(a) Fisher's Method
#95% confidence interval for (LC,SC)
LB=np.tanh(np.arctanh(r1))-1.96/np.sqrt(len(applicant1)-3)
UB=np.tanh(np.arctanh(r1))+1.96/np.sqrt(len(applicant1)-3)
Fisher_CI1=[LB,UB]
Fisher_CI1
```

```python
#95% confidence interval for (SC,SMS)
LB=np.tanh(np.arctanh(r2))-1.96/np.sqrt(len(applicant1)-3)
UB=np.tanh(np.arctanh(r2))+1.96/np.sqrt(len(applicant1)-3)
Fisher_CI2=[LB,UB]
Fisher_CI2
```

```python
#95% confidence interval for (LC,SMS)
LB=np.tanh(np.arctanh(r3))-1.96/np.sqrt(len(applicant1)-3)
UB=np.tanh(np.arctanh(r3))+1.96/np.sqrt(len(applicant1)-3)
Fisher_CI3=[LB,UB]
Fisher_CI3
```

```python
#(b) Ruben's Method
def Ruben(n,u,r):
    r_star=r/np.sqrt(1-r**2)
    a=2*n-3-u**2
    b=r_star*np.sqrt((2*n-3)*(2*n-5))
    c=(2*n-5-u**2)*r_star**2-2*u**2
    
    y1=(2*b-np.sqrt(4*b**2-4*a*c))/(2*a)
    y2=(2*b+np.sqrt(4*b**2-4*a*c))/(2*a)
    
    LB=y1/np.sqrt(1+y1**2)
    UB=y2/np.sqrt(1+y2**2)
    CI=[LB,UB]
    
    return CI
```

```python
#95% confidence interval for (LC,SC)
Ruben_CI1=Ruben(len(applicant1),1.96,r1)
Ruben_CI1
```

```python
#95% confidence interval for (SC,SMS)
Ruben_CI2=Ruben(len(applicant1),1.96,r2)
Ruben_CI2
```

```python
#95% confidence interval for (LC,SMS)
Ruben_CI3=Ruben(len(applicant1),1.96,r3)
Ruben_CI3
```

```python
#4
#Use mean substitution method for missing values in all variables except 'rep78'
auto3=auto.reindex(columns=['model','origin', 'price', 'mpg', 'rep77', 'hroom', 'rseat', 
                'trunk', 'weight', 'length', 'turn', 'displa', 'gratio'])
auto3
```

```python
mean_sub=auto.fillna(value=auto3.mean())
mean_sub
```

```python
import statsmodels.formula.api as sm
```

```python
#Use regression method
model = sm.ols(formula="rep78 ~ price + mpg + rep77 + hroom + rseat + trunk + weight + length + turn + displa + gratio", data=mean_sub).fit()
model.params
```

```python
#rows with NaN for variable 'rep78'
missing=mean_sub.loc[[2,9,52,56,62],:]
missing
```

```python
prediction=pd.DataFrame(model.predict(missing), columns=['rep78'])
prediction
```

```python
#fill missing values with predicted values
regression = mean_sub.fillna(value=prediction)
regression
```

```python
#Mean and standard deviation before imputation
print(auto['rep78'].mean(),auto['rep78'].std())
```

```python
#Mean and standard deviation after imputation
print(regression['rep78'].mean(),regression['rep78'].std())
```
